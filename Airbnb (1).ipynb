## 3. ***Data Wrangling***

### Data Wrangling Code

````python
# Write your code to make your dataset analysis ready.

# Remove duplicates
df = df.drop_duplicates()
````

````python
# Drop rows with missing host_name or name (very few, negligible impact)
df = df.dropna(subset=['host_name', 'name'])

````

````python
# Handle missing values
df['reviews_per_month'] = df['reviews_per_month'].fillna(0)
df['last_review'] = pd.to_datetime(df['last_review'], errors='coerce')
````

````python
# Cap extreme prices
price_cap = df['price'].quantile(0.99)
df['price'] = np.where(df['price'] > price_cap, price_cap, df['price'])
````

````python
# Cap extreme outliers in minimum_nights (reasonable threshold, e.g., 365)
df['minimum_nights'] = np.where(df['minimum_nights'] > 365, 365, df['minimum_nights'])
````

````python
# Ensure availability is within valid range (0–365)
df['availability_365'] = df['availability_365'].clip(0, 365)
````

````python
# Price per night (normalize by minimum nights)
df['price_per_night'] = df['price'] / df['minimum_nights']

````

````python
# Helpful derived fields
df["is_shared"] = df["room_type"].eq("Shared room").astype(int)
df["is_private"] = df["room_type"].eq("Private room").astype(int)
df["is_entire"] = df["room_type"].eq("Entire home/apt").astype(int)
````
